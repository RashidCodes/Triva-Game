## TRIVIA GAME
A simple trivia game for any number of players. Each player gets a turn at answering 5 trivia questions. When a question is displayed, 4 possible answers are also displayed. Only one of the answers is correct, and if the player selects the correct answer, he or she earns a point. 


After answers have been selected per question, the program displays the number of points earned by the player, the player with the highest number of points wins.


Just coding for fun here :).

```python
## Let us create our Question class
class Question:
    def __init__(self, question, ans_1, ans_2, ans_3, ans_4, correct_ans):
        self.question = question
        self.ans_1 = ans_1
        self.ans_2 = ans_2
        self.ans_3 = ans_3
        self.ans_4 = ans_4
        self.correct_ans = correct_ans
        
    
    def __repr__(self):
        return "Question(question= {})".format(self.question)
        
```

There all done, let us create a list of all questions

```python
## 10 questions
question_1 = Question(
    question = 'What is the color of my cap?',
    ans_1 = 'green',
    ans_2 = 'blue',
    ans_3 = 'red',
    ans_4 = 'black',
    correct_ans = 'green'
)

question_2 = Question(
    question = "How old are you?",
    ans_1 = '10',
    ans_2 = '12',
    ans_3 = '21',
    ans_4 = '24',
    correct_ans = '21'
)

question_3 = Question(
    question = "What is the color of your mac",
    ans_1 = "white",
    ans_2 = 'grey',
    ans_3 = 'black',
    ans_4 = 'indigo',
    correct_ans = 'grey'
)

question_4 = Question(
    question = 'What day is today',
    ans_1 = 'Monday',
    ans_2 = 'Tuesday',
    ans_3 = 'Wednesday',
    ans_4 = 'Thursday',
    correct_ans = 'Wednesday'
)

question_5 = Question(
    question = "In which region was the independence day of 2019 celebrated",
    ans_1 = "Northern",
    ans_2 = "Southern",
    ans_3 = "Easter",
    ans_4 = "Greater-Accra",
    correct_ans = "Northern"
)

question_6 = Question(
    question = "Who is the current president of the republic",
    ans_1 = "Mohammed Rashid",
    ans_2 = 'Rashid Mohammed',
    ans_3 = 'Nana Akuffo Addo',
    ans_4 = 'Akuffo Addo Nana Addo Danquah',
    correct_ans = 'Akuffo Addo Nana Addo Danquah'
)

question_7 = Question(
    question = 'What is my height',
    ans_1 = "5'2",
    ans_2 = "5'7",
    ans_3 = "6'2",
    ans_4 = "6'3",
    correct_ans = "6'3"
)

question_8 = Question(
    question = "What is the first line of the national anthem",
    ans_1 = "I promise on my Honour",
    ans_2 = "God Bless our Homeland",
    ans_3 = "To be faithful and loyal to Ghana my Motherland",
    ans_4 = "God bless our Homeland Ghana",
    correct_ans = "God bless our Homeland Ghana"
)

question_9 = Question(
    question = "A Monday born from the Akan ethnic group is called..?",
    ans_1 = "Kofi",
    ans_2 = 'Kwadwo',
    ans_3 = 'Yaw',
    ans_4 = 'Kwasi',
    correct_ans = 'Kwadwo'
)

question_10 = Question(
    question = "How many questions are you supposed to answer?",
    ans_1 = '5',
    ans_2 = '4',
    ans_3 = '3',
    ans_4 = '2',
    correct_ans = '5'
)

## Let's keep up with the score of the player
score = 0

## Let's now create a list of our 10 questions
list_ = [question_1, question_2, question_3, question_4, question_5, question_6, question_7, question_8,question_9, question_10]


## generating random sets of 5 questions.
from itertools import combinations
random_questions = [combos for i, combos in enumerate(combinations(list_, 5))]

## we have 252 sets of 5 questions, all we need to do is generate a random integer 0<= integer <=252.
from random import randint
integer_ = randint(0, 253)

def main(list_):
    ## generate a random integer for the range 0<= x <= 252
    integer_ = randint(0, 252)
    
    ## select our combo of questions
    questions_ = list(list_[integer_])
    
    ## Now let's write a program that'll display the question for each question class, and take an answer from the user.
    def main_within(question):
        global score
        print('QUESTION: ')
        print(question.question)
        print('\n')
        print('ANSWERS:')
        print('*',question.ans_1)
        print('*',question.ans_2)
        print('*',question.ans_3)
        print('*',question.ans_4)
        correct_ans = question.correct_ans
        print('\n')

        answer = input("Please enter your answer: ")
        print('\n')

        ## increase the score variable by 1 for each correct answer
        if answer == correct_ans:
            score += 1
            print('SCORE:', score)
        
        else:
            print('SCORE: ', score)
            
    # call the main_within function for each question       
    for question in questions_:
        main_within(question)
        print('\n')
        
    if score == 5:
        print('Congratulations on your perfect score')
        
```

```python
## run the program
main(random_questions)
```
